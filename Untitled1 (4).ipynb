Data Transformation
I have prepared this Jupyter notebook to give a brief introduction to data transformations. See Referenes for sources and further details. The first part of the notebook describes theory while the second part shows a synthetic example in Python.

Theory
Data transformation is the application of a deterministic mathematical function to each point in a data set1. This function should be invertible and continuous. Main aims of the transformation are:

Achieve numerical stability, e.g., in regression
Achieve linear separability in otherwise non-linear data set
Reduce variance and thus dimensions, e.g., PCA
To obtain certain, after normal, distribution
Comparability, e.g., unit conversion
Comprehensive visualization
If possible, always visualize data prior and after the transformation

There are different types of transformation. These can be divided (in simplified way) into:

Linear transformation
Linear function with offset and multiplication (scale) that can be applied successively.
Can be presented as coordinate transformation composed of a shift of the origin and axis scaling (e.g., via projection)

Pros

It does not break-up linear relations between dependent and independent variable
Simple and mostly computationally inexpensive
User is on a safe side when applying linear transformation

Cons

Will not help to solve non-linear separability or visualization problems

Examples3

MinMaxScaler: scale to specific renge
StandardScaler: subtract mean and set to unit standard deviation
Normalizer: normalize the data to have unit norm, independently of the distribution of the samples (norm can be maximum values, abs. sum (L1) square sum (L2))
Refers to a per sample transformation instead of a per feature transformation
RobustScaler: similar to StandardScaler but using only 1st-3rd quartile to compute median and deviation that are then used for whole data set

Is linear, does not remove outliers

PCA: is ortogonal = linear tranformation of the coordinate system providing all components are preserved
Requires uses of StandardScaler = zero mean and 1 variance

Non-linear transformation
Can have arbitrary form, offset, multiplication, and other variable scale

Pros
Transformed data can be used to solve non-linear problems
Useful for modeling issues related to heteroscedasticity (non-constant variance)
Can be applied to meet criteria for certain distribution and thus to apply common rules
For example 2*sigma rule for normal distribution

Cons
breaks up linear relations
For example of square or log transformation: the difference of 0.1 between 1.0 and 1.1 is after transformation equal to difference between 2.0 and 2.1
Following blindly the aim to meet some distribution after transformation does not change the fact that the original data set will still not follow such distribution
Furthermore, such assumptions may lead to false and literally dangerous conclusions

Examples3
QuantileTransformer: transforms the data to follow a uniform or a normal distribution.
Can handle multi-nomial distributions
This transformation tends to spread out the most frequent values
It also reduces the impact of (marginal) outliers: this is therefore a robust preprocessing scheme
Because it is a non-parametric method, it is harder to interpret than the parametric ones (power)7

PowerTransformer: a parametric, monotonic transformations that is applied to make data more Gaussian-like.

Scikit-learn supports Box-Cox6 (non-negative values only) and the Yeo-Johnson6 transformation
The power transform finds the optimal scaling factor to stabilize variance and mimimize skewness through maximum likelihood estimation3

KernelPCA: non-linear dimensionality reduction through the use of kernels


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer
from sklearn.preprocessing import PowerTransformer, QuantileTransformer
```

```python
%config InlineBackend.figure_format = "svg"
```

```python
#Create data set with non-zero mean, variance, including outliers
x0, y0 = np.meshgrid(np.linspace(0.0,200.0,num=30), np.linspace(-100,100,num=20))
# Convert to feature matrix
X = np.concatenate((np.asanyarray(x0).reshape(x0.size,1),
                    np.asanyarray(y0).reshape(y0.size,1)),axis=1)
y = 10.0+3.0*X[:,0]-2.0*X[:,1]+0.1*(X[:,0]**2)+0.4*(X[:,1]**2)
np.random.seed(123)

# insert outlier
y[0:20] = y[0:20] + np.random.rand(20)*1e+4

np.random.shuffle(y)
y = y.reshape(y.size,1)
```

```python
#Function to plot scaling result
def plotscaled(vec,i,ncol=2,nrow=6,title=""):
    plt.subplot(nrow,ncol,i)
    i+=1
    plt.hist(vec,24)
    plt.legend([title])
    plt.subplot(nrow,ncol,i)
    plt.plot(vec,"k.")
    y_mean,y_med,y_std,y_min,y_max = np.mean(vec),np.median(vec),np.std(vec),np.min(vec),np.max(vec)
    plt.plot([0,vec.size],[y_mean,y_mean],"r-")
    plt.plot([0,vec.size],[y_med,y_med],"g--")
    plt.plot([0,vec.size],[y_mean-y_std*2,y_mean-y_std*2],"b-")
    plt.plot([0,vec.size],[y_mean+y_std*2,y_mean+y_std*2],"b-")
    if i == 2:
        plt.legend(["data","mean","median","2*$\sigma$"],ncol=4,fontsize=9);
    return i+1
```

```python
plt.figure(figsize=(9.5,9))
i = 1
for m in [StandardScaler(),RobustScaler(),MinMaxScaler(),
          PowerTransformer(),QuantileTransformer(output_distribution="normal")]:
    y_scaled = m.fit_transform(y);
    i = plotscaled(y_scaled,i,title=str(m).split("(")[0])
```

User-defined non-linear transformation
In most cases, either positive or negative skewness should be corrected
Left-skewed (negative) distribution/data can be often "normalized" using log transformation
Righ-skewed (positive) distribution/data can be sometimes "normalized" using square/cube root transformation

```python
from scipy.stats import skewnorm
```

```python
def plotskew(s):
    y = skewnorm.rvs(s, size=1000)
    plt.figure(figsize=(8,5))
    plt.subplot(2,2,1)
    plt.hist(y,20)
    if s > 0:
        plt.title("Positive skew")
    else:
        plt.title("Negative skew")
    plt.subplot(2,2,2)
    return y
```

```python
np.random.seed(1)
y = plotskew(4)
plt.hist(np.log(y-np.min(y)+1),20);
plt.title("log-transform: log(const+y)")
y = plotskew(-4)
plt.hist((y-np.min(y))**2,20)
plt.title("square power-transform: (const+y)^2")
```
